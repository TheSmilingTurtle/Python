```python
%matplotlib widget
import matplotlib.pyplot as plt
```

```python
x = [0, 1, 2, 3, 4, 5, 6, 7]
y = [0.1, 0.7, 2.1, 2.8, 3.7, 5.3, 6, 7.2]
mult = [x*y for x, y in zip(x,y)]
xi_sq = [a**2 for a in x]

assert(len(x) == len(y))
```

```python
w0 = (sum(x)*sum(mult)-sum(xi_sq)*sum(y)) / (sum(x)**2 - len(x)*sum(xi_sq))
w1 = sum(y) / sum(x) - len(x) * (sum(x)*sum(mult) - sum(xi_sq)*sum(y)) / (sum(x)**3 - len(x)*sum(xi_sq)*sum(x))

print(f"{w0} + {w1}x")
```

```python
plt.scatter(x,y)
plt.plot([0,7], [w0, w0 + w1*7])
plt.show()
```
