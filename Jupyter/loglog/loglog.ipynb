```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
```

```python
%%timeit
x = np.linspace(0, 100, 100)
mask1 = np.array([True]*30 + [False]*70)
mask2 = np.array([False]*60 + [True]*40)
y = np.linspace(0, 100, 100)
y = y-25
y[mask1] = 5
y[mask2] = 35
```

```python
plt.plot(x, y)
plt.show()
```
